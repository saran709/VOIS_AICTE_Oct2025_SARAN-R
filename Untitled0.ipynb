```python
# Install dependencies if needed
!pip install openpyxl matplotlib seaborn

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load Excel file
file_path = "/content/1730285881-Airbnb_Open_Data.xlsx"  # Upload to Colab
df = pd.read_excel(file_path, sheet_name="in")

# Basic info
print(df.shape)
df.head()
```

```python
# Check missing values
df.isnull().sum()

# Drop irrelevant columns (like 'license' since almost empty)
df = df.drop(columns=["license"])

# Fill missing text fields with 'Unknown'
df["host name"].fillna("Unknown", inplace=True)
df["neighbourhood"].fillna("Unknown", inplace=True)

# Fill numerical missing with median
df["price"].fillna(df["price"].median(), inplace=True)
df["minimum nights"].fillna(df["minimum nights"].median(), inplace=True)

```

```python
# Summary statistics
df.describe()

# Room type distribution
df["room type"].value_counts()

# Average price per room type
df.groupby("room type")["price"].mean().sort_values(ascending=False)
```

```python
# Listings count per neighbourhood group
plt.figure(figsize=(8,5))
sns.countplot(data=df, x="neighbourhood group", order=df["neighbourhood group"].value_counts().index)
plt.title("Listings by Neighbourhood Group")
plt.show()

# Top 10 neighbourhoods by listings
top_neigh = df["neighbourhood"].value_counts().head(10)
top_neigh.plot(kind="barh", figsize=(8,5), color="skyblue")
plt.title("Top 10 Neighbourhoods by Listings")
plt.show()
```

```python
# Distribution of price (remove extreme outliers)
plt.figure(figsize=(8,5))
sns.histplot(df[df["price"] < 500]["price"], bins=50, kde=True)
plt.title("Price Distribution (below $500)")
plt.show()

# Price vs Room Type
plt.figure(figsize=(8,5))
sns.boxplot(data=df[df["price"] < 500], x="room type", y="price")
plt.title("Price by Room Type")
plt.show()
```

```python
# Top 10 hosts by number of listings
top_hosts = df["host id"].value_counts().head(10)
top_hosts.plot(kind="bar", figsize=(8,5), color="orange")
plt.title("Top 10 Hosts by Listings")
plt.show()

# Verified vs Unverified
sns.countplot(data=df, x="host_identity_verified")
plt.title("Host Identity Verification Status")
plt.show()
```

```python
# Reviews distribution
plt.figure(figsize=(8,5))
sns.histplot(df["number of reviews"], bins=50, kde=False)
plt.xlim(0,200)  # focus on 0-200 reviews
plt.title("Number of Reviews Distribution")
plt.show()

# Availability distribution
plt.figure(figsize=(8,5))
sns.histplot(df["availability 365"], bins=30)
plt.title("Availability (Days per Year)")
plt.show()
```

```python
# Correlation between numerical features
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()

```

```python
# Scatter plot of listings on map
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x="long", y="lat", hue="neighbourhood group", alpha=0.5)
plt.title("Geographical Distribution of Listings")
plt.show()
```
